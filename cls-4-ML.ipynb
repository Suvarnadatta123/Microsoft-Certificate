```python
# Linear Regression (Price Prediction)
from sklearn.linear_model import LinearRegression
import numpy as np

# Input (X) and Output (Y)
X = np.array([[1000], [1500], [2000], [2500]])  # size
y = np.array([5, 7, 9, 11])  # price

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict
print(model.predict([[1800]]))
```

```python
# Classification (Spam Detection Concept)
from sklearn.tree import DecisionTreeClassifier

# Example data
X = [[1, 20], [2, 30], [3, 40], [4, 50]]  # features
y = [0, 0, 1, 1]  # 0 = Not Spam, 1 = Spam

model = DecisionTreeClassifier()
model.fit(X, y)

print(model.predict([[3, 35]]))
```

```python
# Unsupervised Learning Example
# K-Means Clustering (Customer Segmentation)
import os
os.environ["OMP_NUM_THREADS"] = "1"

from sklearn.cluster import KMeans
import numpy as np

# Customer data (income, spending)
X = np.array([[10, 20], [15, 25], [30, 40], [35, 45]])

model = KMeans(n_clusters=2)

model.fit(X)

print(model.labels_)   # Cluster groups
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load
df = pd.read_csv("data.csv")

# Split
X = df[["capital-gain"]]
y = df["income"]

X_train, X_test, y_train, y_test = train_test_split(X, y)

# Train
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
print(model.predict(X_test))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "year": [2020, 2021, 2022],
    "sales": [100, 150, 200]
})

df.plot(x="year", y="sales", kind="line")
plt.show()
df.plot(x="year", y="sales", kind="bar")
plt.show()
```

```python
df["sales"].plot(kind="hist")
plt.show()
```

```python
df.set_index("year")["sales"].plot(kind="pie")
plt.show()
```

```python
import pandas as pd

data = pd.read_csv("data.csv", encoding='latin1')
arr = data.to_numpy()

print(arr[:5])
```

```python
# ARRAY OPERATIONS
print(arr.shape)    # (rows, columns)
print(arr.ndim)     # dimensions
print(arr.size)     # total elements
print(arr.dtype)    # data type
```

```python
# INDEXING AND SLICING
arr[0]        # first row
arr[:, 0]     # first column
arr[0:5]      # first 5 rows
arr[1:3, 0:2] # sub-matrix
```

```python
# MATHIMATICAL OPERATIONS
arr + 10
arr - 5
arr * 2
arr / 2
arr ** 2
```

```python
# Statistical Operations
np.mean(arr)
np.median(arr)
np.std(arr)
np.var(arr)
np.sum(arr)
np.min(arr)
np.max(arr)
```

```python
np.mean(arr)
```

```python
np.mean(arr, axis=0)
```

```python
# Boolean Operations (Filtering)
arr > 50
arr[arr > 50]
```

```python
# Conditional Selection
np.where(arr > 50, 1, 0)
```

```python
# Sorting
np.sort(arr, axis=0)   # column-wise
np.sort(arr, axis=1)   # row-wise
```

```python
# Aggregation
np.sum(arr, axis=0)
np.max(arr, axis=1)
```

```python
# Reshaping
arr.reshape(2, -1)
arr.flatten()
```

```python
np.delete(arr, 0, axis=0)   # delete row
np.delete(arr, 0, axis=1)   # delete column
```

```python
# Unique & Count
np.unique(arr)
```

```python
# Matrix Operations
arr.T                # transpose
np.dot(arr, arr.T)   # matrix multiplication
```

```python
# Handling Missing Values
np.isnan(arr)
np.nan_to_num(arr)
```

```python
# Logical Operations
np.logical_and(arr > 10, arr < 50)
```

```python
# Stacking
np.vstack((arr, arr))
np.hstack((arr, arr))
```

```python
# Splitting
np.split(arr, 2)
```

```python
# Save Back to CSV
np.savetxt("output.csv", arr, delimiter=",")
```
